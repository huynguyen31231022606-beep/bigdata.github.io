```python
# Install libraries if needed
!pip install requests pandas --quiet

import requests
import pandas as pd
from functools import reduce

# ‚úÖ Country - Only Vietnam
country = "VNM"

# ‚úÖ Indicators required (Updated with IIP + Export & Import)
indicators = {
    "GDP Growth (%)": "NY.GDP.MKTP.KD.ZG",
    "CPI (Inflation, annual %)": "FP.CPI.TOTL.ZG",
    "Exchange Rate (USD/VND)": "PA.NUS.FCRF",
    "Interest Rate (%) (Lending Rate Proxy)": "FR.INR.LEND",
    "FDI Inflow (US$)": "BX.KLT.DINV.CD.WD",
    "Exports (% of GDP)": "NE.EXP.GNFS.ZS",
    "Imports (% of GDP)": "NE.IMP.GNFS.ZS"
}

start_year = 2010
end_year = 2024

# ‚úÖ World Bank API data fetch function
def get_wb_data(country_code, indicator_code, start, end):
    url = (
        f"https://api.worldbank.org/v2/country/{country_code}/indicator/"
        f"{indicator_code}?format=json&date={start}:{end}&per_page=2000"
    )
    r = requests.get(url)
    data = r.json()

    # N·∫øu kh√¥ng c√≥ d·ªØ li·ªáu
    if len(data) < 2 or not isinstance(data[1], list) or len(data[1]) == 0:
        print(f"‚ö†Ô∏è Warning: No data for indicator {indicator_code}")
        return pd.DataFrame()

    df = pd.DataFrame(data[1])[["date", "value"]]
    df["Year"] = df["date"].astype(int)
    df.drop(columns=["date"], inplace=True)
    df.rename(columns={"value": indicator_code}, inplace=True)
    return df

# ‚úÖ Retrieve all indicators for Vietnam
data_frames = []
for name, code in indicators.items():
    print(f"Fetching {name} ...")
    df_temp = get_wb_data(country, code, start_year, end_year)
    if df_temp.empty:
        continue  # b·ªè DataFrame r·ªóng
    df_temp.rename(columns={code: name}, inplace=True)
    data_frames.append(df_temp)

# ‚úÖ Merge into a single table, ch·ªâ khi c√≥ DataFrame t·ªìn t·∫°i
if data_frames:
    df = reduce(lambda left, right: pd.merge(left, right, on="Year", how="outer"), data_frames)
    df = df.sort_values("Year").reset_index(drop=True)
else:
    df = pd.DataFrame()
    print("‚ùå No data available for any indicator!")

# ‚úÖ Formatting display
pd.options.display.float_format = '{:,.2f}'.format

# ‚úÖ Export Excel n·∫øu c√≥ d·ªØ li·ªáu
if not df.empty:
    output_filename = "Vietnam_Macro_Indicators_2010_2024.xlsx"
    df.to_excel(output_filename, index=False)
    print("‚úÖ File exported successfully:", output_filename)
else:
    print("‚ùå No Excel file exported, data empty.")

df
```

```python
# ============================================================
# üßπ STEP 1: Chu·∫©n h√≥a gi√° c·ªï phi·∫øu ng√†nh th√©p (daily)
# ============================================================
df_gia_thep['time'] = pd.to_datetime(df_gia_thep['time'])
df_gia_thep = df_gia_thep.sort_values('time')
df_gia_thep.rename(columns={'time': 'Date', 'close': 'Stock_Close'}, inplace=True)

# ============================================================
# ‚öôÔ∏è STEP 2: Chu·∫©n h√≥a d·ªØ li·ªáu gi√° nguy√™n li·ªáu (monthly)
# ============================================================
# Gi·∫£ ƒë·ªãnh 'Ng√†y' l√† d·∫°ng chu·ªói ng√†y ƒë·∫ßu th√°ng ho·∫∑c cu·ªëi th√°ng
price_raw.rename(columns={'Ng√†y': 'Date'}, inplace=True)
price_raw['Date'] = pd.to_datetime(price_raw['Date'], errors='coerce')
price_raw = price_raw[['Date', 'L·∫ßn cu·ªëi']].rename(columns={'L·∫ßn cu·ªëi': 'Input_Price'})
price_raw = price_raw.dropna(subset=['Date'])

# ============================================================
# üí∞ STEP 3: Chu·∫©n h√≥a b√°o c√°o t√†i ch√≠nh (quarterly)
# ============================================================
finance = steel_finance.copy()
finance = finance.rename(columns={'CP': 'Symbol', 'NƒÉm': 'Year', 'K·ª≥': 'Quarter'})
# Chuy·ªÉn 'NƒÉm' v√† 'K·ª≥' th√†nh c·ªôt th·ªùi gian th·ª±c
finance['Quarter'] = finance['Quarter'].astype(str).str.replace('Q', '')
finance['Quarter'] = pd.to_numeric(finance['Quarter'], errors='coerce')
finance['Date'] = pd.to_datetime(finance['Year'].astype(str) + '-' + (finance['Quarter']*3).astype(int).astype(str) + '-01', errors='coerce')
finance = finance.dropna(subset=['Date'])
finance = finance[['Symbol', 'Date', 'T·ªîNG C·ªòNG T√ÄI S·∫¢N (ƒë·ªìng)', 'V·ªêN CH·ª¶ S·ªû H·ªÆU (ƒë·ªìng)', 'N·ª¢ PH·∫¢I TR·∫¢ (ƒë·ªìng)']]

# ============================================================
# üåè STEP 4: Chu·∫©n h√≥a ch·ªâ s·ªë kinh t·∫ø vƒ© m√¥ (annual)
# ============================================================
econ = df_econ.copy()
econ['Year'] = econ['Year'].astype(int)
econ['Date'] = pd.to_datetime(econ['Year'].astype(str) + '-12-31')
econ = econ[['Date', 'GDP Growth (%)', 'CPI (Inflation, annual %)', 
             'Exchange Rate (USD/VND)', 'Interest Rate (%) (Lending Rate Proxy)', 
             'FDI Inflow (US$)']]

# ============================================================
# üîó STEP 5: Merge d·ªØ li·ªáu v·ªõi ƒëi·ªÅu ch·ªânh th·ªùi gian
# ============================================================

# --- 5.1: Merge gi√° th√©p daily + nguy√™n li·ªáu monthly ---
df_merged = pd.merge_asof(
    df_gia_thep.sort_values('Date'),
    price_raw.sort_values('Date'),
    on='Date',
    direction='backward'
)

# --- 5.2: ƒêi·ªÅu ch·ªânh d·ªØ li·ªáu t√†i ch√≠nh sang qu√Ω sau ---
finance_adj = finance.copy()
finance_adj['Date'] = finance_adj['Date'] + pd.offsets.QuarterEnd(1)  # d·ªùi sang cu·ªëi qu√Ω ti·∫øp theo

df_merged = pd.merge_asof(
    df_merged.sort_values('Date'),
    finance_adj.sort_values('Date'),
    on='Date',
    direction='backward'
)

# --- 5.3: ƒêi·ªÅu ch·ªânh d·ªØ li·ªáu kinh t·∫ø vƒ© m√¥ sang nƒÉm sau ---
econ_adj = econ.copy()
econ_adj['Date'] = econ_adj['Date'] + pd.offsets.YearEnd(1)  # d·ªùi sang cu·ªëi nƒÉm ti·∫øp theo

df_merged = pd.merge_asof(
    df_merged.sort_values('Date'),
    econ_adj.sort_values('Date'),
    on='Date',
    direction='backward'
)

# ============================================================
# üì¶ STEP 6: Xu·∫•t d·ªØ li·ªáu t·ªïng h·ª£p
# ============================================================
output_file = "steel_data_ready_for_ML.xlsx"
df_merged.to_excel(output_file, index=False)

print("‚úÖ Data merged successfully!")
print(f"üìÅ Saved as: {output_file}")
display(df_merged.head(10))
```

```python
df_merged['Stock_Return'] = df_merged['Stock_Close'].pct_change()
df_merged['Cumulative_Return'] = (1 + df_merged['Stock_Return']).cumprod()
df_merged['Drawdown'] = df_merged['Cumulative_Return'] / df_merged['Cumulative_Return'].cummax() - 1

fig, axs = plt.subplots(2,1,figsize=(12,7), sharex=True)
axs[0].plot(df_merged.index, df_merged['Cumulative_Return'], color='green')
axs[0].set_title("Cumulative Returns")
axs[1].plot(df_merged.index, df_merged['Drawdown'], color='red')
axs[1].set_title("Drawdown")
plt.show()
```

```python
!pip install mplfinance --quiet
import mplfinance as mpf
import pandas as pd

# ƒê·∫£m b·∫£o index l√† Date
df_plot = df_merged[['Stock_Open','Stock_High','Stock_Low','Stock_Close','Stock_Volume']].copy()
df_plot.index.name = 'Date'

# rename c·ªôt theo chu·∫©n mplfinance
df_plot.rename(columns={
    'Stock_Open':'Open',
    'Stock_High':'High',
    'Stock_Low':'Low',
    'Stock_Close':'Close',
    'Stock_Volume':'Volume'
}, inplace=True)

# Moving Average
df_plot['MA20'] = df_plot['Close'].rolling(20).mean()
df_plot['MA50'] = df_plot['Close'].rolling(50).mean()

# Bollinger Bands
df_plot['BB_Mid'] = df_plot['MA20']
df_plot['BB_Upper'] = df_plot['MA20'] + 2*df_plot['Close'].rolling(20).std()
df_plot['BB_Lower'] = df_plot['MA20'] - 2*df_plot['Close'].rolling(20).std()

# MACD
ema12 = df_plot['Close'].ewm(span=12, adjust=False).mean()
ema26 = df_plot['Close'].ewm(span=26, adjust=False).mean()
df_plot['MACD'] = ema12 - ema26
df_plot['Signal'] = df_plot['MACD'].ewm(span=9, adjust=False).mean()

# Ch·ªâ v·∫Ω ph·∫ßn d·ªØ li·ªáu m·ªõi nh·∫•t (v√≠ d·ª• 200 ng√†y) ƒë·ªÉ r√µ
df_plot_recent = df_plot[-200:]

apds = [
    mpf.make_addplot(df_plot_recent['MA20'], color='blue'),
    mpf.make_addplot(df_plot_recent['MA50'], color='orange'),
    mpf.make_addplot(df_plot_recent['BB_Upper'], color='grey', linestyle='--'),
    mpf.make_addplot(df_plot_recent['BB_Lower'], color='grey', linestyle='--'),
    mpf.make_addplot(df_plot_recent['MACD'], panel=1, color='green', ylabel='MACD'),
    mpf.make_addplot(df_plot_recent['Signal'], panel=1, color='red')
]

mpf.plot(
    df_plot_recent,
    type='candle',
    style='charles',
    addplot=apds,
    volume=True,
    title='Steel Stock - Candlestick with MA, Bollinger Bands & MACD',
    figsize=(14,8),
    panel_ratios=(3,1)
)
```
